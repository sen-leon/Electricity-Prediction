```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n_param=4
# import forecast data
parameters_fc = ["wind_speed", "sunshine_duration", "pressure_air_site_reduced", "temperature_air_mean_200", "cloud_cover_total"]
parameter = parameters_fc[n_param]
path="../data/weather_forecasts/"+parameter+".pkl"
df_fc=pd.read_pickle(path)

#import historical data
parameters_hist = ["wind_speed", "sunshine_duration", "pressure_air_site", "temperature_air_mean_200", "cloud_cover_total"]
parameter = parameters_hist[n_param]
path="../data/weather/"+parameter+"_2021.pkl"
df_hist = pd.read_pickle(path)

df_fc = df_fc.set_index(df_fc.date).drop(["date", "timestamp"], axis=1)
df_hist = df_hist.set_index(df_hist.date).drop(["date", "timestamp"], axis=1)
```

print("2021 (historical):")
print(df_fc["2022-09-29":"2022-09-30"])

print("2022 (forecast):")
print(df_hist["2021-09-29":"2021-09-30"])

```python
plt.figure()
fig = plt.gcf()
fig.set_size_inches(20, 6.5)
stations = df_hist.columns
station = stations[5]
print("station_id: ", station)
index=df_fc.loc["2022-09-29":"2022-10-08", station].index
plt.plot(index, df_fc.loc["2022-09-29":"2022-10-08", stations], lw=3, label="Forecast (2022)")
plt.plot(index, df_hist.loc["2021-09-29":"2021-10-08", station], lw=3, label="Historical (2021)")
plt.legend(loc='upper left')
#plt.xlim([850, 980])
#plt.ylim([0,1])

plt.show()
```
